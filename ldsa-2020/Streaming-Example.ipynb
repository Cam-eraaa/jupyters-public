```python
# Based on https://github.com/apache/spark/blob/master/examples/src/main/python/streaming/network_wordcount.py

from pyspark.sql import SparkSession
from pyspark import SparkContext
from pyspark.streaming import StreamingContext

# New API
spark_session = SparkSession\
        .builder\
        .master("spark://ben-uppmax-haste-spark-master:7077") \
        .appName("PythonStreamingWordcount")\
        .config("spark.dynamicAllocation.enabled", True)\
        .config("spark.shuffle.service.enabled", True)\
        .config("spark.dynamicAllocation.executorIdleTimeout","30s")\
        .config("spark.executor.cores",4)\
        .getOrCreate()

BATCH_INTERVAL_SECONDS = 5

```

```python
# We'll use Spark Streaming/DStreams -- the 'old API'

spark_streaming_context = StreamingContext(spark_session.sparkContext, BATCH_INTERVAL_SECONDS)

lines = spark_streaming_context.socketTextStream('stream-src-1', 9999)

lines.flatMap(lambda line: line.split(' '))\
     .map(lambda word: (word,1))\
     .reduceByKey(lambda a, b: a + b).pprint()


```

```python
spark_streaming_context.start()

spark_streaming_context.awaitTermination()
```
